#### Notes
Explanations for some of the fields can be found at: https://docs.snomed.org/snomed-ct-specifications/snomed-ct-release-file-specification/appendices/appendix-e-concept-enumerations  
1. ModuleId : present in all SNOMED files. Refers to development module
2. DefinitionStatusId : Found in concepts.txt. Only 2 choices: 9xxx...73002 (fully defined) and 9xxx...74008 (incompletely defined; concept is considered to be primitive.
3. Description TypeId : Found in description.txt. Only 3 choices: 9xxx...3001 (fully-specified), 9xxx13009 (synonym in the language indicated by LanguageCode field in description.txt), 9xxx...50004 (textual definition of the associated concept in the language indcated by LanguageCode field).
4. CaseSignificanceId : Found in description.txt and text_definition.txt. 3 choices: 9xxx...17005 (text must be presented in the case in which it is specified), 9xxx...20002 (case-insensitive for first letter in the term), 9xxx...448009 (entire description term is case-insensitive and can be changed to upper or lowercase). Probably not useful.
5. CharacteristicTypeId : Found in relationship.txt, stated_relationship.txt and relationship_concrete_values.txt. Refers to relationship. Only 1 entry, so not useful. Should remove during cleaning.
6. ModifierId : also found in the 3 relationships files. only 1 choice, so can also drop
 

```python
datasets_old = [concepts, describe, relate, define, state, concrete, owl]
```

```python
# what are the values for active and languageCode?

active = []

for dataset in datasets_old:
    unique_active = list(dataset['active'].unique())
    active.append(unique_active)

active_set = set([item for sublist in active for item in sublist])
active_set
```

```python
# slice out the inactive entries
concepts, describe, relate, define, state, concrete, owl = [df[df['active'] == 1] for df in datasets_old]
```

```python
# languageCode
lang = []

for dataset in datasets_old:
    if 'languageCode' in dataset.columns:
        unique_lang = list(dataset['languageCode'].unique())
        lang.append(unique_lang)

lang_set = set([item for sublist in lang for item in sublist])
lang_set
# set contains only english, so can remove languageCode column
```

```python
# Cleaning

to_remove = ['caseSignificanceId', 
             'active',
             'characteristicTypeId',
             'modifierId',
             'effectiveTime',
             'moduleId',
             'definitionStatusId',
             'languageCode']


        
```

```python
# do for the rest
for dataframe in datasets_old:
    remove(dataframe)
```

```python
datasets_new = [df.drop(columns=[col for col in to_remove if col in df.columns]) for df in datasets_old]
concepts, describe, relate, define, state, concrete, owl = datasets_new
```

```python
# save to csv
csv_names = {'concept': concepts,
             'description': describe,
             'relationship': relate,
             'text_definition': define,
             'stated_relationship': state,
             'relationship_concrete_values': concrete,
             'refset_owl_expression': owl}

for filename_prefix, dataframe in csv_names.items():
    dataframe.to_csv(filename_prefix + '_cleaned.csv', index=False)
```
